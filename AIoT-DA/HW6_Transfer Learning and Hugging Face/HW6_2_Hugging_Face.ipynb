步驟 2：加載 Stable Diffusion 模型

```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
import matplotlib.pyplot as plt

# 加載 Stable Diffusion 2.0 模型
pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1")
pipe.to("cuda")  # 如果有 GPU，加速模型運行

# 函數：根據描述生成圖像
def generate_image_from_text(text_prompt):
    image = pipe(text_prompt).images[0]

    # 使用 matplotlib 顯示圖片
    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis("off")  # 隱藏座標軸
    plt.show()

    return image
```

步驟 3：手動輸入文字並生成圖像

```python
# 手動輸入描述並生成圖像
text_input = input("請輸入您想要生成的圖像描述: ")
generated_image = generate_image_from_text(text_input)
```
